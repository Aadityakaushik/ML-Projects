```python
import polars as pl
```

```python
cdf=pl.read_csv("D:\jupyter\dataset\calories.csv")
edf=pl.read_csv("D:\jupyter\dataset\exercise.csv")
```

```python
cdf=cdf.drop(columns="User_ID")
edf=edf.drop(columns="User_ID")
```

```python
cdf.head()
```

```python
edf.head()
```

```python
cdf.describe()
```

```python
edf.describe()
```

```python
df=pl.concat([ edf,cdf,],how="horizontal")
```

```python
df.head()
```

```python
from_=['male','female']
to=[1,0]
```

```python
df=df.with_columns(
    pl.when(pl.col('Gender')==from_[0])
    .then(to[0])
    .when(pl.col('Gender')==from_[1])
    .then(to[1])
    .alias('gender')
)
df=df.drop(columns='Gender')
```
