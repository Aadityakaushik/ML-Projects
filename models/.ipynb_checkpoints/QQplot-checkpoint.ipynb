```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
```

```python
data=load_iris()
```

```python
X=data.data
```

```python
data.feature_names
```

```python
X=pd.DataFrame(X,columns=data.feature_names)
```

```python
X
```

```python
temp=sorted(X['sepal length (cm)'].tolist())
```

```python
quantiles=[]
for i in range(1,101):
    quantiles.append(np.percentile(temp,i))
```

```python
theoretical_data=np.random.normal(loc=0,scale=1,size=100)
```

```python
theo_quant=[]
for i in range(1,101):
    theo_quant.append(np.percentile(theoretical_data,i))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(quantiles,theo_quant)
```

```python
import statsmodels.api as sm
```

```python
fig=sm.qqplot(X['sepal length (cm)'],line='45',fit=True)
plt.title('qq plot')
```

```python
uniform_data=np.random.uniform(low=0,high=1,size=1000)
```

```python
plt.hist(uniform_data)
```

```python
import scipy.stats as stats
```

```python
params=stats.uniform.fit(uniform_data)
dist=stats.uniform(loc=params[0],scale=params[1])
```

```python
fig=sm.qqplot(uniform_data,dist=dist,line='45')
plt.title('qq plot of uniform data with uniform fit')
plt.show()
```
