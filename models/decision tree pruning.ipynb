```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
```

```python
data=load_iris()
data
```

```python
x=data.data
y=data.target
```

```python
x.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtr,xte,ytr,yte=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
xtr.shape,xte.shape,ytr.shape,yte.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier() #defalt is gini
```

```python
model.fit(xtr,ytr)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(model,filled=True)
```

```python
pruned_model=DecisionTreeClassifier(max_depth=2)
pruned_model.fit(xtr,ytr)
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(pruned_model,filled=True)
```

```python
y_pred=model.predict(xte)
y_pred
```

```python
y_pruned_pred=pruned_model.predict(xte)
y_pruned_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(yte,y_pred),accuracy_score(yte,y_pruned_pred)
```

```python
# pre pruning
```

```python
parameters={
    'criterion':['gini','entropy','log_loss'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto','sqrt','log2']
    
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
model3=DecisionTreeClassifier()
cv=GridSearchCV(model3,param_grid=parameters,cv=5,scoring='accuracy')
```

```python
cv.fit(xtr,ytr)
```

```python
cv.best_params_
```

```python
y_pred=cv.predict(xte)
accuracy_score(yte,y_pred)
```
