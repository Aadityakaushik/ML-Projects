```python
import pandas as pd
import numpy as np
import scipy.stats as stats
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import FunctionTransformer
from sklearn.compose import ColumnTransformer
```

```python
df=pd.read_csv('/home/aaditya/system/vs/jupyter/dataset/titanic.csv')
df.head()
```

```python
df.info()
```

```python
df.drop()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
x=df[['Age','Fare']]
y=df['Survived']
```

```python
x_tr,x_te,y_tr,y_te=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.displot(x_tr['Age'])
plt.title('Age PDF')
plt.show()

stats.probplot(x_tr['Age'],dist='norm',plot=plt)
plt.title('Age QQ plot')
```

```python
sns.displot(x_tr['Fare'])
plt.title('Fare PDF')
plt.show()

stats.probplot(x_tr['Fare'],dist='norm',plot=plt)
plt.title('Fare QQ plot')

#right skewed
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
clf1=LogisticRegression()
clf2=DecisionTreeClassifier()
```

```python
clf1.fit(x_tr,y_tr)
clf2.fit(x_tr,y_tr)

print(f'accuracy for LR : {accuracy_score(y_te,clf1.predict(x_te))}')
print(f'accuracy for DT : {accuracy_score(y_te,clf2.predict(x_te))}')
```

```python
trf = FunctionTransformer(func=np.log1p)
```

```python
X_train_transformed = trf.fit_transform(x_tr)
X_test_transformed = trf.transform(x_te)
```

```python
clf = LogisticRegression()
clf2 = DecisionTreeClassifier()

clf.fit(X_train_transformed,y_tr)
clf2.fit(X_train_transformed,y_tr)
    
y_pred = clf.predict(X_test_transformed)
y_pred1 = clf2.predict(X_test_transformed)
    
print("Accuracy LR",accuracy_score(y_te,y_pred))
print("Accuracy DT",accuracy_score(y_te,y_pred1))
```

```python
X_transformed = trf.fit_transform(x)

clf = LogisticRegression()
clf2 = DecisionTreeClassifier()

print("LR",np.mean(cross_val_score(clf,X_transformed,y,scoring='accuracy',cv=10)))
print("DT",np.mean(cross_val_score(clf2,X_transformed,y,scoring='accuracy',cv=10)))
```

```python
plt.figure(figsize=(14,4))

plt.subplot(121)
stats.probplot(x_tr['Fare'], dist="norm", plot=plt)
plt.title('Fare Before Log')

plt.subplot(122)
stats.probplot(X_train_transformed['Fare'], dist="norm", plot=plt)
plt.title('Fare After Log')

plt.show()
```

```python
plt.figure(figsize=(14,4))

plt.subplot(121)
stats.probplot(x_tr['Age'], dist="norm", plot=plt)
plt.title('Age Before Log')

plt.subplot(122)
stats.probplot(X_train_transformed['Age'], dist="norm", plot=plt)
plt.title('Age After Log')

plt.show()
```

```python
trf2 = ColumnTransformer([('log',FunctionTransformer(np.log1p),['Fare'])],remainder='passthrough')

X_train_transformed2 = trf2.fit_transform(x_tr)
X_test_transformed2 = trf2.transform(x_te)
```

```python
clf = LogisticRegression()
clf2 = DecisionTreeClassifier()

clf.fit(X_train_transformed2,y_tr)
clf2.fit(X_train_transformed2,y_tr)
    
y_pred = clf.predict(X_test_transformed2)
y_pred2 = clf2.predict(X_test_transformed2)
    
print("Accuracy LR",accuracy_score(y_te,y_pred))
print("Accuracy DT",accuracy_score(y_te,y_pred2))
```

```python
X_transformed2 = trf2.fit_transform(x)

clf = LogisticRegression()
clf2 = DecisionTreeClassifier()

print("LR",np.mean(cross_val_score(clf,X_transformed2,y,scoring='accuracy',cv=10)))
print("DT",np.mean(cross_val_score(clf2,X_transformed2,y,scoring='accuracy',cv=10)))
```
